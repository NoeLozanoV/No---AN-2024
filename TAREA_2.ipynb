```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt
```

```python
x,y,z = sy.symbols('x y z')
x = sy.symbols('x')
f = sy.sin(x)+2*x
f1 = sy.series(f,x,-0.2,22)
f2 = sy.series(f,x,-0.1,22)
f3 = sy.series(f,x,0.1,22)
f4 = sy.series(f,x,0.2,22)
```

```python
f1O= f1.removeO()
f2O= f2.removeO()
f3O= f3.removeO()
f4O= f4.removeO()
```

```python
lamf = sy.lambdify(x,f)
lamf1 = sy.lambdify(x,f1O)
lamf2 = sy.lambdify(x,f2O)
lamf3 = sy.lambdify(x,f3O)
lamf4 = sy.lambdify(x,f4O)
```
