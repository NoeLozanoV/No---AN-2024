```python
import numpy as np
import sympy as sy
from sympy import *
import matplotlib.pyplot as plt
```

```python
x=sy.symbols('x')

#Método de la cuerda

def cuerda(x,s):
    C=x-(2*x**2)/s
    print(C)
    if C==x:
        return C
    else:
        return cuerda(C,s)
```

```python
x=1
s=2
cuerda(x,s)
```

```python
#Método de Newton

def newton(y,x,s):
    N=x-(2*x**2)/(4*x)
    y=y+1
    print(N)
    if N== x:
        print("Número de iteraciones realizadas:",y)
        return N
    else:
        return newton(y,N,s)
```

```python
y=0
x=2
s=3
newton(y,x,s)
```
